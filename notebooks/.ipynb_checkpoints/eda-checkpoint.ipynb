## ðŸ“Š Exploratory Data Analysis (EDA)

This notebook covers the initial data analysis steps for the customer churn dataset. It includes:
- Data cleaning and preprocessing
- Handling missing values and encoding categorical features
- Visualization of feature distributions and correlations
- Insights into churn vs. non-churn customers

Key Takeaway: Identified important features potentially influencing churn, guiding further modeling efforts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset from the data folder
df = pd.read_csv('D:/datascienceprojects/customerchurnp/data/data.csv')

# Display the first 5 rows
df.head()
```

```python
# Get data types and non-null counts
df.info()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

- There are no missing values
- Total charges shows to be an object type but it should be float type

```python
# Try converting TotalCharges to numeric, coerce errors into NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Now check how many got converted to NaN
df['TotalCharges'].isnull().sum()
```

```python
# Drop rows where TotalCharges is NaN
df = df[df['TotalCharges'].notnull()]

# Reset index (optional but cleaner)
df.reset_index(drop=True, inplace=True)

# Confirm no more NaNs
df['TotalCharges'].isnull().sum()
```

```python
# Value counts (absolute)
df['Churn'].value_counts()
```

```python
# Value counts (percentage)
df['Churn'].value_counts(normalize=True) * 100
```

- The data is imbalanced as the percentages of no churn and churn are so far apart.

---

```python
# Exclude ID column
df = df.drop(columns=['customerID'])

# Separate columns by data type
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Remove target from categorical (if present)
categorical_cols.remove('Churn')

print("Categorical columns:", categorical_cols)
print("Numerical columns:", numerical_cols)
```

```python
categorical_cols
```

```python
# Set plot style
sns.set(style="whitegrid")

# Set figure size
plt.figure(figsize=(20, 25))

# Loop through all categorical features
for i, col in enumerate(categorical_cols):
    plt.subplot(5, 3, i + 1)
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f"{col} vs Churn", fontsize=15)
    
plt.tight_layout()
plt.show()
```

- Gender does not seem to be a significant feature
- Partners and Dependents seem to be highly correlated to the churn rate
- The rest of the features seem to show slight trends as well

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df, x=col, hue='Churn', kde=True, multiple='stack', palette='Set2')
    plt.title(f"{col} Distribution by Churn", fontsize=12)
    plt.tight_layout()

plt.show()
```

```python
# Encode churn as 1 (Yes) and 0 (No) to use in correlation
df_corr = df.copy()
df_corr['Churn'] = df_corr['Churn'].map({'No': 0, 'Yes': 1})

# Correlation with target
print(df_corr.corr(numeric_only=True)['Churn'].sort_values(ascending=False))
```

- Tenure and TotalCharges seem to be negatively correlated to the churn rate

```python
# Boxplots
plt.figure(figsize=(18, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, hue='Churn', y=col, palette='Set3')
    plt.title(f"{col} Boxplot by Churn", fontsize=12)
    plt.tight_layout()

plt.show()
```

```python
categorical_cols
```

```python
from sklearn.preprocessing import LabelEncoder

# Copy dataframe to avoid overwriting
df_encoded = df.copy()

# Binary categorical columns (can use LabelEncoder)
binary_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

le = LabelEncoder()
for col in binary_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# One-hot encode other categorical columns
df_encoded = pd.get_dummies(df_encoded, columns=[
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaymentMethod'
])

df_encoded.head()
```

```python
print("Label Encoded Mappings:")
for col in df_encoded:
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f"{col}: {mapping}")

# 2. Visualize one-hot encoded columns (grouped)
print("\nOne-Hot Encoded Columns:")
one_hot_cols = [
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaymentMethod'
]

# Count the number of dummy variables created per feature
dummy_counts = {col: len([c for c in df_encoded.columns if c.startswith(col + "_")]) for col in one_hot_cols}

# Plot it
plt.figure(figsize=(10, 5))
sns.barplot(hue=list(dummy_counts.keys()), y=list(dummy_counts.values()), palette='viridis')
plt.xticks(rotation=45)
plt.title("Number of One-Hot Encoded Columns Per Feature")
plt.ylabel("Number of Dummy Columns")
plt.tight_layout()
plt.show()
```

```python
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Drop 'customerID' (not useful for prediction)
df_model = df_encoded.copy()

# 2. Separate label (target)
X = df_model.drop('Churn', axis=1)
y = df_model['Churn']

# 3. Train-test split (stratify = preserve churn balance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 4. Scale numerical columns only
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

scaler = StandardScaler()

# Fit only on training data to prevent data leakage
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

```python
import joblib

# Save train-test split
joblib.dump(X_train, 'D:/datascienceprojects/customerchurnp/data/X_train.pkl')
joblib.dump(X_test, 'D:/datascienceprojects/customerchurnp/data/X_test.pkl')
joblib.dump(y_train, 'D:/datascienceprojects/customerchurnp/data/y_train.pkl')
joblib.dump(y_test, 'D:/datascienceprojects/customerchurnp/data/y_test.pkl')
```
