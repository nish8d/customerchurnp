This notebook implements and evaluates three traditional classifiers:
- Logistic Regression
- Random Forest
- XGBoost
Each model is:
- Trained and evaluated using accuracy, precision, recall, and F1 score
- Threshold-tuned for improved performance
- Saved using `joblib` for later use

```python
import joblib

# Load preprocessed data
X_train = joblib.load('../data/X_train.pkl')
X_test = joblib.load('../data/X_test.pkl')
y_train = joblib.load('../data/y_train.pkl')
y_test = joblib.load('../data/y_test.pkl')
```

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Initialize and train the model
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
log_reg.fit(X_train, y_train)

# Predictions
y_pred_lr = log_reg.predict(X_test)

# Evaluation
print("Logistic Regression Classification Report:\n")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(6,4))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Logistic Regression Confusion Matrix')
plt.show()
```

```python
# Initialize and train
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)

# Predictions
y_pred_rf = rf.predict(X_test)

# Evaluation
print("Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(6,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
# Initialize and train
xgb = XGBClassifier(eval_metric='logloss', scale_pos_weight=1.8, random_state=42)
xgb.fit(X_train, y_train)

# Predictions
y_pred_xgb = xgb.predict(X_test)

# Evaluation
print("XGBoost Classification Report:\n")
print(classification_report(y_test, y_pred_xgb))

# Confusion Matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(6,4))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Oranges', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('XGBoost Confusion Matrix')
plt.show()
```

---

```python
thresholds = np.arange(0.0, 1.01, 0.01)

# Get probabilities
y_probs_xgb = xgb.predict_proba(X_test)[:, 1]
y_probs_rf = rf.predict_proba(X_test)[:, 1]
y_probs_logreg = log_reg.predict_proba(X_test)[:, 1]

# Calculate F1 scores
f1_xgb = [f1_score(y_test, y_probs_xgb > t) for t in thresholds]
f1_rf = [f1_score(y_test, y_probs_rf > t) for t in thresholds]
f1_logreg = [f1_score(y_test, y_probs_logreg > t) for t in thresholds]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f1_xgb, label='XGBoost', color='orange')
plt.plot(thresholds, f1_rf, label='Random Forest', color='green')
plt.plot(thresholds, f1_logreg, label='Logistic Regression', color='blue')

# Highlight best points
best_thresh_xgb = thresholds[np.argmax(f1_xgb)]
best_thresh_rf = thresholds[np.argmax(f1_rf)]
best_thresh_logreg = thresholds[np.argmax(f1_logreg)]

plt.axvline(best_thresh_xgb, color='orange', linestyle='--', alpha=0.5)
plt.axvline(best_thresh_rf, color='green', linestyle='--', alpha=0.5)
plt.axvline(best_thresh_logreg, color='blue', linestyle='--', alpha=0.5)

# Labels
plt.title('F1-Score vs Threshold')
plt.xlabel('Threshold')
plt.ylabel('F1-Score')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Predict probabilities
y_probs_xgb = xgb.predict_proba(X_test)[:, 1]

# Find best threshold
thresholds = np.arange(0.0, 1.01, 0.01)
f1_scores = [f1_score(y_test, y_probs_xgb > t) for t in thresholds]
best_threshold_xgb = thresholds[np.argmax(f1_scores)]

print(f"Best Threshold (XGBoost based on F1): {best_threshold_xgb:.2f}")

# Apply best threshold
y_pred_thresh_xgb = (y_probs_xgb > best_threshold_xgb).astype(int)

# Evaluation
print("Classification Report (After Threshold Tuning - XGBoost):")
print(classification_report(y_test, y_pred_thresh_xgb))
```

```python
# Predict probabilities
y_probs_rf = rf.predict_proba(X_test)[:, 1]

# Find best threshold
f1_scores_rf = [f1_score(y_test, y_probs_rf > t) for t in thresholds]
best_threshold_rf = thresholds[np.argmax(f1_scores_rf)]

print(f"Best Threshold (Random Forest based on F1): {best_threshold_rf:.2f}")

# Apply best threshold
y_pred_thresh_rf = (y_probs_rf > best_threshold_rf).astype(int)

# Evaluation
print("Classification Report (After Threshold Tuning - Random Forest):")
print(classification_report(y_test, y_pred_thresh_rf))
```
